# Exploratory Data Analysis

### Exploratory Data Analysis is the process of exploring data, generating insights, testing hypotheses, checking assumptions and revealing underlying hidden patterns in the data. This process typically makes use of descriptive statistics and visualizations.

```python
import requests
import bs4
```

```python
def url_to_transcript(url):
    res = requests.get(url)
    data = bs4.BeautifulSoup(res.text,'lxml')
    text = [p.text for p in data.find(class_ = "elementor-element elementor-element-74af9a5b elementor-widget elementor-widget-theme-post-content").find_all('p')]
    print(url)
    return text
```

# RegEx Module

### A RegEx, or Regular Expression, is a sequence of characters that forms a search pattern. RegEx can be used to check if a string contains the specified search pattern.



```python
import re
import string
def clean_text_round1(text):
    text = text.lower()
    text = re.sub('\[.*?\]','',text)
    text = re.sub('[%s]'%re.escape(string.punctuation),'',text)
    text = re.sub('\w*\d\w*','',text)
    return text
```

```python
round1 = lambda x: clean_text_round1(x)
```

```python
data_clean = pd.DataFrame(data_df.transcript.apply(round1))
data_clean
```

```python
def clean_text_round2(text):
    text = re.sub('[‘’“”...]','',text)
    text = re.sub('\n','',text)
    return text
```

```python
round2 = lambda x: clean_text_round2(x)
```

```python
data_clean = pd.DataFrame(data_clean.transcript.apply(round2))
data_clean
```

# WordCloud Module
### Word Cloud is a data visualization technique used for representing text data in which the size of each word indicates its frequency or importance

```python
from wordcloud import WordCloud
```

```python
wc = WordCloud(stopwords = stop_words, background_color = "white", colormap = "Dark2",
              max_font_size = 150, random_state = 42)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=[16,6]
full_names = ['Vir Das','Ali Wong', 'Anthony Jeselnik', 'Bill Burr',
             'Bo Burnham', 'Dave Chappelle', 'Hasan Minhaj', 'Jim Jefferies',
             'Joe Rogan', 'John Mulaney', 'Mike Birbiglia', 'Ricky Gervais']

for index,comedian in enumerate(data.columns):
    #print(index)
    wc.generate(data_clean.transcript[comedian])
    plt.subplot(3,4,index+1)
    plt.imshow(wc)
    plt.axis('off')
    plt.title(full_names[index])
plt.show
```

# TextBlob Module
### TextBlob is a Python (2 and 3) library for processing textual data. TextBlob can be used to perform a variety of NLP tasks ranging from parts-of-speech tagging to sentiment analysis, and language translation to text classification.

```python
from textblob import TextBlob
```

```python
pol = lambda x: TextBlob(x).sentiment.polarity
sub = lambda x: TextBlob(x).sentiment.subjectivity

data_clean['polarity'] = data_clean['transcript'].apply(pol)
data_clean['subjectivity'] = data_clean['transcript'].apply(sub)
data_clean['full_names']= full_names
```

```python
data_clean
```

```python
plt.rcParams['figure.figsize']=[10,8]
for i,comedian in enumerate(data_clean.index):
    x = data_clean.polarity.loc[comedian]
    y = data_clean.subjectivity.loc[comedian]
    plt.scatter(x,y,color='red')
    plt.text(x+0.001, y+0.001, full_names[i], fontsize=10)
    plt.xlim(-0.01,0.12)
    

plt.title('Sentiment Analysis', fontsize=20)
plt.xlabel('<--- Negative ------ Positive --->', fontsize=15)
plt.ylabel('<--- Facts ------ Opinions --->', fontsize=15)
plt.show()
```
